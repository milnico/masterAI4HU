```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets

n_samples = 100
X,y = datasets.make_regression(n_samples = n_samples,n_features = 1,noise=5)

plt.scatter(X,y)
w = np.random.random()
b = np.random.random()
epochs = 100
lr = 0.1
for i in range(epochs):
  for id, x in enumerate(X):
    y_pred = w*x + b
    err = y_pred-y[id]

    w = w - lr*err*x
    b = b - lr*err*1

y_final = w*X + b
plt.plot(X,y_final,'r')
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from tensorflow.keras import layers
import tensorflow as tf


n_samples = 500
x_data = np.linspace(-10,10,num=n_samples)
y_data = 0.1*x_data*np.cos(x_data) + 0.1*np.random.normal(size=n_samples)

model = tf.keras.Sequential()
model.add(tf.keras.Input(shape=(1,)))#primo layer neuroni di input
model.add(layers.Dense(32,use_bias=True,activation="tanh"))#secondo layer neuroni di hidden
model.add(layers.Dense(16,use_bias=True,activation="tanh"))#terzo layer neuroni di hidden
model.add(layers.Dense(1,use_bias=True,activation="linear"))
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
model.compile(loss="mean_squared_error",optimizer=opt)
model.fit(x_data,y_data,epochs=100,batch_size=32,verbose=1)
print(np.shape(x_data))
prediction = model(x_data.reshape((len(x_data),1))).numpy()

plt.scatter(x_data,y_data)

plt.plot(x_data,prediction.reshape((len(prediction,))),'r')
```
