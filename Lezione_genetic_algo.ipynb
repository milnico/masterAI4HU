```python
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_der(x):
    return x*(1-x)

def step_func(z):
        return np.where(z<0.5,0,1)

class NN:
    def __init__(self,inputs):
        self.inputs = inputs
        self.h =4#len(inputs)
        self.li = len(inputs[0])
        self.wi = (2*np.random.random((self.li,self.h)))-1
        self.wh = (2*np.random.random((self.h,1)))-1
        self.lr = 1
        
    def predict(self,inp):
        s1=sigmoid(np.dot(inp,self.wi))
        s2=sigmoid(np.dot(s1,self.wh))
        return s2


    def train(self,inputs,outputs,it):
        for i in range(it):
            l0= inputs
            
            l1=sigmoid(np.dot(l0,self.wi))
            
            l2 = sigmoid(np.dot(l1,self.wh))
            
            
            l2_err=outputs-l2
            #print l2_err
            l2_delta=np.multiply(l2_err,sigmoid_der(l2))

            l1_err = np.dot(l2_delta,self.wh.T)
            l1_delta = np.multiply(l1_err,sigmoid_der(l1))

            self.wh += self.lr*np.dot(l1.T,l2_delta)
            self.wi += self.lr*np.dot(l0.T,l1_delta)

inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
outputs = np.array([[0],[1],[1],[0]])
net = NN(inputs)
print("Before")
print(net.predict(inputs))
net.train(inputs,outputs,10000)
print("After ")
print(net.predict(inputs))
```

```python
import numpy as np



def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_der(x):
    return x*(1-x)

# Sorting function


# genetic algorithm
def evo_algo():
    
    mutRate=0.05
    #netInitialization
    inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
    outputs = np.array([[0], [1], [1], [0]])
    net = NN(inputs, outputs)
    
    ngenes = net.netParameter()
    maxevaluations = 100
    findSolution = "false"

    # setting parameters
    popsize = 20  # population size, offspring number
    
   
    genome = np.arange(ngenes*popsize*2, dtype=np.float64).reshape((popsize*2,ngenes))
    # allocate and initialize the solution center
    for i in range(popsize*2):
			  genome[i,:] = (np.random.random((1,ngenes))*2.0)-1
        
    cgen=0

    # main loop
    while (cgen < maxevaluations):
        cgen = cgen + 1

        fitness = np.zeros(popsize*2)
        for k in range(popsize*2):
            fitness[k] = net.evaluateNet(genome[k]) #+ (np.random.random()*2.0 - 1)*0.2
            
            

        # Sort by fitness 
      
        
        index = np.argsort(fitness)  # maximization
        #(index)
        #print(fitness[index])
        genome = genome[index]
        #print(genome)
        for p in range(popsize):
          for g in range(ngenes):
            if np.random.random() > mutRate:
              genome[p,g] = genome[p+popsize,g]
            else:
              genome[p,g] = np.random.random()*2.0 - 1
        

        bestfit = fitness[-1]
        if bestfit == 1.0:
            findSolution = "true"
        
        print(' Gen %d Bestfit %.2f' % (
                cgen, bestfit))

    print("After ")
    print(net.predict(genome[k]))


class NN:
    def __init__(self,inputs,outputs):
        self.inputs = inputs
        self.outputs = outputs
        self.l =20#len(inputs) hidden neurons number
        self.li = len(inputs[0])
        self.wi = np.random.random((self.li,self.l))
        self.wh = np.random.random((self.l,1))


    def netParameter(self):
        return self.li*self.l + self.l*1

    def train(self,inputs,outputs,it):
        for i in range(it):
            l0= inputs
            l1=sigmoid(np.dot(l0,self.wi))

            l2 = sigmoid(np.dot(l1,self.wh))
            l2_err=outputs-l2
            #print l2_err
            l2_delta=np.multiply(l2_err,sigmoid_der(l2))

            l1_err = np.dot(l2_delta,self.wh.T)
            l1_delta = np.multiply(l1_err,sigmoid_der(l1))

            self.wh += np.dot(l1.T,l2_delta)
            self.wi += np.dot(l0.T,l1_delta)

    def predict(self,genotype):
        l0 = self.inputs
        first_layer = self.li*self.l
        self.wi = genotype[:first_layer].reshape(self.li,self.l)
        second_layer = self.l*1
        self.wh =genotype[first_layer:].reshape(self.l,1)
        l1 = np.tanh(np.dot(l0, self.wi))
        l1 = np.where(l1<0.5,0,1)
        l2 = np.dot(l1, self.wh)
        #print(l2)
        l2 = np.where(l2<0.5,0,1)
        
        return l2

    def evaluateNet(self,genotype):
        netOut = self.predict(genotype)
        #print(self.outputs)
        #print(netOut)
        fitness = np.sum(netOut==self.outputs)
        #fitness = np.mean(np.abs((netOut-self.outputs)))
        #print(fitness)
        #input("hee#")
        return fitness


evo_algo()
```
