## Vector

A vector is a list of number ( can be in a row or column). A vector of length n is just a sequence of n numbers. In numpy, we have arrays method which takes an object as a parameter.

```python
import numpy as np
Vr = np.array([[1, 2, 3]]) # row vector
Vc = np.array([[1],[2], [3]]) # Column vector
## To find out the shape of a vector 
print(np.shape(Vr))#Vr.shape # output => (1, 3)
print(np.shape(Vc))#Vc.shape # output => (3, 1)
```

Addition of two vectors, we add them element-by-element


```python
x = np.ones(3)       # Vector of three ones
y = np.array((5, 4, 6))   # Converts tuple (5, 4, 6) into array
x + y                  #output => array([8, 7, 9])
#Similarly, subtraction of two vectors
y - x                 # output => array([2, 1, 3]) 
```

Multiply each element of a vector by a scalar

```python
4 * x     # output => array([4, 4, 4])
5 * y     # output => array([ 25, 20, 30])
```

# The dot product
The dot product,prodotto scalare ,of two vectors is a scalar. Matrix multiplication is the dot product of two matrices.

dot_prod.png

```python
import numpy as np
x = np.array([1,2,3])
y = np.array([4,5,6])
print(np.shape(x),np.shape(y))
np.dot(x,y)                 # output= 32
```

# Hadamard Product
element-wise multiplication and outputs a vector

hadamard_product.png

```python
X = np.array([2, 3, 4])
Y = np.array([5,6, 7])
H = X * Y           # 2 *5, 3* 6, 4*7
print(H)             #output = [10, 18, 28]
```

# Matrix multiplication
Matrix multiplication relies on dot product to multiply various combination of row and column.
Rules for matrix multiplication

The number of columns of the 1st matrix must equal the number of rows of the 2nd.

The product of an M x N matrix and an N x K matrix is an M x K matrix. The new matrix takes the rows of the 1st and columns of the 2nd.

VK2Qtby.jpg

```python
import numpy as np
a = np.array([[1, 0], [0, 1]])
b = np.array([[4, 1], [2, 2]])
print(a)
print(b)
np.dot(a, b)
```

```python
import numpy as np
A = np.array([[1, 2,3], [3,2,1], [3,2,1]])
print(np.shape(A))
B = np.array([[1, 2], [3, 2],[2, 7]])
print(np.shape(B))
np.dot(A,B)
```

Note that using the * for multiplication will result in hadamard product or elementwise operation.matrix_had.png

# Transpose

Numpy method .transpose() can be applied to any numpy array to get the transpose matrix.

Transpose.gif

```python
A = np.array([[1, 2],[3,4],[5, 6]])
print(A)
print(np.shape(A))
AT = A.transpose()  #Give the transpose
print(np.shape(AT))
#shorter notation
A.T                 #Give the transpose
```
